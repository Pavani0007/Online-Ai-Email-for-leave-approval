```python
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM
import torch

import ipywidgets as widgets
from IPython.display import display, Markdown
from datetime import datetime, timedelta

# Use an instruction-tuned model (like Zephyr or Falcon)
model_name = "tiiuae/falcon-7b-instruct"  # or "HuggingFaceH4/zephyr-7b-beta"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.float16, device_map="auto")

letter_generator = pipeline("text-generation", model=model, tokenizer=tokenizer)

# Now use your generate_leave_letter function as before


def generate_leave_letter(employee_name, reason, days, recipient="HR Manager"):
    start_date = datetime.now().strftime("%d %B %Y")
    end_date = (datetime.now() + timedelta(days=days)).strftime("%d %B %Y")

    prompt = (
        f"You are a helpful assistant that writes professional leave application letters.\n"
        f"Please draft a formal leave application addressed to the {recipient}.\n\n"
        f"Employee Name: {employee_name}\n"
        f"Reason for Leave: {reason}\n"
        f"Duration of Leave: {days} days (from {start_date} to {end_date})\n\n"
        f"Sincerely: {employee_name}\n"
        f"Write the letter in a professional and polite tone. Do not invent or hallucinate any details.\n"
        f"Start the letter below:\n\n"
        f"Letter:\n"
    )

    output = letter_generator(
        prompt,
        max_length=3000,
        num_return_sequences=1,
        temperature=0.7,
        truncation=True,
        pad_token_id=50256  # This helps avoid weird cutoff
    )[0]['generated_text']

    # Clean result (remove the prompt from output if repeated)
    letter = output.split("Letter:")[-1].strip()
    return letter

# Widgets
name_input = widgets.Text(description="Your Name:")
reason_input = widgets.Text(description="Reason:")
days_input = widgets.IntSlider(description="Days:", min=1, max=30)
recipient_input = widgets.Text(description="To:", value="HR Manager")
generate_btn = widgets.Button(description="Generate Letter")
output = widgets.Output()

def on_generate_click(b):
    with output:
        output.clear_output()
        letter = generate_leave_letter(
            name_input.value,
            reason_input.value,
            days_input.value,
            recipient_input.value
        )
        display(Markdown(f"**Generated Leave Letter:**\n\n{letter}"))

generate_btn.on_click(on_generate_click)

display(widgets.VBox([
    name_input,
    reason_input,
    days_input,
    recipient_input,
    generate_btn,
    output
]))
```
